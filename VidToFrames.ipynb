```python
import cv2
import time
import os

# function to convert one video file into number of frames == length of video (seconds)
def video_to_frames(input_loc, output_loc, filename):
    count = 0
    vidcap = cv2.VideoCapture(input_loc)
    success,image = vidcap.read()
    success = True
    while success:
        try:
            vidcap.set(cv2.CAP_PROP_POS_MSEC,(count*1000))    # added this line 
            success,image = vidcap.read()
            print ('Read a new frame: ', success)
            cv2.imwrite(output_loc + f"\\frame{filename}{count}.jpg", image)     # save frame as JPEG file
            count = count + 1
        except:
            print("Done")
```

```python
# function to convert all videos in a directory into frames. Saved at "output_dir"
def dir_to_frames(input_dir, output_dir):
    for filename in os.listdir(input_dir):
        input_loc = os.path.join(input_dir,filename)
        vid_name = os.path.basename(filename).split(sep = ".")[0]
        video_to_frames(input_loc, output_dir, vid_name)
```

```python
input_dir = "./FakeImages"
output_dir = "./ProcessedFakeImages"

dir_to_frames(input_dir, output_dir)
```
