## Image Formatter

- Resize images down to a standard set size, rename and convert to jpg format.
- Images will be numbered from 1 to n where n is the number of input images.
- Provide address to target input folder and target output folder.
- Current dimension is 720 Horizontal and 1280 Vertical to retain the semi rectangle shape

```python
import os
from PIL import Image

# Reads in all images in a folder as long as image format
def format_images(input_folder, new_horz, new_vert, output_folder):
    image_number = 1
    for filename in os.listdir(input_folder):
        image_path = input_folder + '/' + filename
        print('Reading image: ' + image_path)
        img = Image.open(image_path)
        # Resize image
        resized_image = img.resize((new_horz, new_vert))
        rgb_image =  resized_image.convert('RGB')
        new_image_name = output_folder + str(image_number) + '_BunnyLogo' + '.jpg'
        print('Saving image to ' + new_image_name)
        rgb_image.save(new_image_name, 'JPEG')
        image_number += 1
    print('Done!')
```

```python

# Run script - Green pass
format_images(input_folder="./data/raw_images", new_horz=720, new_vert=1280, output_folder="./data/images/")
# # Format red passes
# format_images(input_folder="./RawImages/RedPass", new_horz=720, new_vert=1280, output_folder="./ProcessedImages/RedPass/")
```
