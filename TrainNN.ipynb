## Notebook for training of Final Output NN

```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from sklearn.model_selection import train_test_split
from keras.layers import Dense
import tensorflow as tf
```

```python
df = pd.read_csv('./trainingSet.csv')
df
```

```python
labels = df['Label']
labels
```

```python
features = df.iloc[:,0:5]
features
```

```python
# Split the training data into training and validation sets
X=features
y=np.ravel(labels)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=50)

y_train = pd.get_dummies(y_train).values
y_test = pd.get_dummies(y_test).values

X_test
```

```python
model = Sequential()
model.add(Dense(5, activation='tanh', input_shape=(5,)))
# model.add(Dense(5, activation='tanh'))
model.add(Dense(3, activation='softmax'))

model_checkpoint = tf.keras.callbacks.ModelCheckpoint("NN.h5", save_best_only=True, monitor='val_accuracy', verbose=1)
model.summary()
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Fit model
model.fit(X_train, y_train, epochs=75, batch_size=8, validation_data=(X_test, y_test), callbacks=[model_checkpoint])
```
